<h2>Identifikasi Data</h2><br>
Langkah pertama yang dilakukan adalah mengenali data, menetapkan variabel tergantung (target) dan variabel bebas (independen), serta memeriksa apakah dalam data tersebut sudah bersih atau masih terdapat data yang hilang.

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Improving Employee Retention by Predicting Employee Attrition Using Machine Learning.xlsx')
df
```

```python
#Menampilkan informasi tipe data setiap kolom
df.info()
```

<h3>1. Missing Value</h3>

```python
#Mendeteksi baris yang kosong
df.isna().sum()
```

```python
# Menghapus baris yang memiliki nilai kosong pada kolom tertentu
df = df.dropna(subset=['SkorKepuasanPegawai', 'JumlahKeikutsertaanProjek', 'JumlahKeterlambatanSebulanTerakhir', 'JumlahKetidakhadiran'])
```

```python
# Mengisi nilai kosong pada kolom 'IkutProgramLOP' dengan 0
df['IkutProgramLOP'] = df['IkutProgramLOP'].fillna(0)
```

```python
# Menghitung nilai mode dari kolom 'AlasanResign'
mode_value = df['AlasanResign'].mode()[0]  # mode() mengembalikan nilai dalam bentuk series, ambil nilai pertama

# Mengisi nilai kosong di kolom 'AlasanResign' dengan nilai mode
df['AlasanResign'] = df['AlasanResign'].fillna(mode_value)
```

```python
df.isna().sum()
```

```python
#Mendeteksi duplikasi baris
df.duplicated().sum()
```

<h3>3. Membuat Kolom Dependen</h3>

```python
# Membuat kolom 'Resign' berdasarkan kolom 'TanggalResign'
df['Resign'] = df['TanggalResign'].apply(lambda x: 1 if pd.notnull(x) and x != '-' else 0)

# Menampilkan data yang sudah diperbarui
print(df[['TanggalResign', 'Resign']])
```

4. Ditermukan baris bernilai kosong pada beberapa kolom:
   * Kolom SkorKepuasanPegawai, JumlahKeikutsertaanProjek, JumlahKeterlambatanSebulanTerakhir, dan JumlahKetidakhadiran memiliki jumlah baris kosong kurang dari 5% jadi akan di dilakukan drop baris.

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

<h3>1. Unvariat Analysis</h3>

```python
# Univariate Analysis menggunakan violin plot

plt.figure(figsize=(10, 15))
for i, col in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols) // 2 + 1, 2, i + 1)  # Adjusting the grid layout for subplots
    sns.violinplot(x='Resign', y=col, data=df, palette="Set2")
    plt.title(f"Violin Plot of {col} by Resign")
    plt.xlabel('Resign')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis menggunakan barplot

plt.figure(figsize=(12, len(categorical_cols) * 5))  # Adjust height based on number of categorical columns

# Loop through each categorical column
for i, col in enumerate(categorical_cols):
    plt.subplot(len(categorical_cols), 1, i + 1)  # One column, multiple rows
    sns.countplot(x=col, hue='Resign', data=df, palette="Set2")
    plt.title(f"Barplot of {col} by Resign")
    plt.xlabel(col)
    plt.ylabel('Count')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis menggunakan density plot

num_cols = len(numeric_cols)
num_rows = (num_cols // 3) + (num_cols % 3 > 0)

plt.figure(figsize=(18, num_rows * 5))  # Adjust figsize for better readability

for i, col in enumerate(numeric_cols):
    plt.subplot(num_rows, 3, i + 1)  # Adjust layout to fit 3 columns
    sns.kdeplot(data=df, x=col, hue='Resign', fill=True, common_norm=False, palette='Set2')
    plt.title(f"Density Plot of {col} by Resign")
    plt.xlabel(col)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
